```python
#plotting using matplotlib

plt.scatter(x,y,c='g')#c defines the color of point

plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title("Graphics in 2D");
```

```python
y = x * x
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,"ro--")
```

```python
plt.plot(x,y,'r*',linestyle='solid',linewidth=5,markersize=12)#markersize can be define using s parameter and use of it to change the marked point
```

```python
# creating multiple subplots
plt.subplot(2,2,1)#first parameter gives nrows, second ->ncolumns ,third->index (position)
plt.plot(a,b,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'y')
plt.subplot(2,2,4)
plt.plot(x,y,'ro')
```

```python
#compute the x and y coordinates for points on a sine curve
x = np.arange(0,4*np.pi,0.05)#firs parmater->start,second->stop,third->step
y = np.sin(x)
plt.title("sine wave form")

# plot the point using matplotlib
plt.plot(x,y)
plt.show();
```

```python
#subplot
#compute the x and y coordinates for points on sine and cosien curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title("sine wave form")
plt.subplot(2,2,2)
plt.plot(x,y_cos,"r")
plt.title('cosine wave form');
plt.subplot(2,2,3);
plt.plot(x,y_tan,'g');
```

```python
#bar plot
x = [2,8,9]
y = [3,4,8]
x2 = [9,8,4]
y2 = [23,1,5]
plt.bar(x,y,align = 'center')
plt.bar(x2,y2,color  ='g')
plt.title('bar graph')
plt.ylabel('Y axis')
plt.xlabel('x axis');
```

### Box Plot using Matplotlib

```python
import random
data = [np.random.normal(0,std,10) for std in range(1,5)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True);#if vert->false then it rotate to horizontal and patch_artist= flase then colors empty on box
```

### pie-chart

```python
#data to plot
labels = ['python','C++','Ruby','Java']
sizes = [250,130,245,210]
colors = ['gold','yellowgreen','red','skyblue']
explode = (0.1,0.1,0.2,0.3)#explode the 1st slide
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct = 'None',shadow=True)#
plt.axis('equal')
plt.show();
```
