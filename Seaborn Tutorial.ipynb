### Correlation with Heatmap
A correlation heaptmap uses coloured cells, typically monochromatic scale to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in featues Selection

```python
df.corr()
```

```python
#let's calculaate total bill of all customers
df.total_bill.sum()
```

```python
df.groupby('sex').total_bill.sum()
```

### Join Plot
A join plot allows the relationship between  2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex');
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pair plot
A 'pair plot' is also known as a scatterplot, in which one variable in the same data row is matched with another varialbes's value, like this: Pairs plots are just elaborations on this, showig all variables paired with all the other variables


```python
sns.pairplot(df)
```

```python
 # counting the number of smoker and non_smoker 
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker');
```

### Displot

### Categorical Plots
Seaborn also helps us in doing the anlaysis on Categroical Data points. In this section we will discuss about

### Countplot
The sns.countplot() function in Seaborn is used to create a bar plot that displays the count of observations in each category of a categorical variable. 

```python

sns.countplot(x='sex',data=df);
```

```python
sns.countplot(data =df,x='smoker')
```

```python
#let's visualize no of customer visit on day basis
sns.countplot(x='day',data=df);
```

from the above bar graph we can visualize the flow of customer is high in saturday,sunday,thursday and friday (which is the lowest flow)

### bar plot

```python
sns.barplot(x='sex',y='total_bill',data=df);
```

### Box plot

```python
sns.boxplot(x='smoker',y='total_bill',data=df);
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow');
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#categorize my data on some others categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

### Violin Plot


Violin Plot helps us to see both the distribution in therms of kernel density estimations and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette ='rainbow');
```
